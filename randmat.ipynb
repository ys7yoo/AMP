```python
import numpy as np
from numpy import linalg as LA
```

```python
r=10
c=100

A=np.random.randn(r,c) / np.sqrt(r) # generate Gaussian random matrix and normalize by the number of row
A=np.matrix(A)
#print(A)
```

```python
AtA = A.T*A
#print(AtA)

import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(AtA)
plt.show()

print(AtA)

diag= np.diag(AtA)
diag.mean()     # indeed diagonal is close to 1 ( a nice property of a random matrix)
# how about histogram?
plt.hist(diag)
```

```python
H = AtA-np.eye(c)
Hf = H.flatten().T
#print(H.flatten().shape)
plt.hist(Hf)
np.var(Hf) # variance ~ 1/r (Lemma 4.2.1, Proof in Appendix E)
```

```python
# norms of rows of H
ns=LA.norm(H,axis=-1)
#ns=np.sqrt(np.sum(np.square(np.abs(H)),axis=-1))
plt.hist(ns)
#ns.mean
```
